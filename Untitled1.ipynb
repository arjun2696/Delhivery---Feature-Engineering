# Processing the data

```python
newdf = df.groupby(["trip_uuid" ,"source_center" ,"source_name"]).agg(
    
    {
        "data" : "min","trip_creation_time":"min", "route_schedule_uuid":"min",
        "route_type" : "min" , "destination_center" : "min" , "destination_name" : "last",
        'od_start_time': "min", 'od_end_time' : "max","start_scan_to_end_scan" : "max",
        "actual_distance_to_destination" : "max", 
        
        
        'actual_time' : "max" , 'osrm_time' : "max", 'osrm_distance' :"max", 'segment_actual_time': "sum",
       'segment_osrm_time' : "sum", 'segment_osrm_distance': "sum"
        
        
    }
)
```

```python
newdf.head()
```

```python
dataset = newdf.reset_index().groupby(["trip_uuid"]).agg(
    {"source_center" : "last" ,"source_name" : "last" , "data" : "min" ,"trip_creation_time" : "min",
     
    "route_schedule_uuid" : "min" , "route_type" : "min" , "destination_center" : "first","destination_name":"first",
     "od_start_time" : "last" , "od_end_time" : "first" , "start_scan_to_end_scan" : "sum" , 
     "actual_distance_to_destination":"sum","actual_time" : "sum" , "osrm_time" : "sum", "osrm_distance" : "sum",
     "segment_actual_time" : "sum" ,  "segment_osrm_time" : "sum" ,  "segment_osrm_distance" : "sum"
   }


)
```

# Feature Engineering

## Adding Features to the Data Set

```python
dataset["Source state"] = dataset.apply(lambda x : x["source_name"].split("(")[1][:-1], axis  = 1)
```

```python
dataset["Source city"] = dataset.apply(lambda x : x["source_name"].split()[0].split("_")[0] , axis = 1)
```

```python
dataset["Destination state"] = dataset.apply(lambda x : x["destination_name"].split("(")[1][:-1], axis= 1)
dataset["Destination city"] = dataset.apply(lambda x : x["destination_name"].split()[0].split("_")[0] , axis = 1)
```

```python
dataset["Trip Creation Year"] = dataset["trip_creation_time"].dt.year
dataset["Trip Creation Month"] = dataset["trip_creation_time"].dt.month
dataset["Trip Creation Day"] = dataset["trip_creation_time"].dt.day
```

```python
dataset["Duration"] = abs(dataset["od_end_time"]-dataset["od_start_time"])
```

```python
dataset["Duration in Minutes"] = dataset["Duration"].dt.total_seconds()/60
```

# Hypthosis Testing

```python
from scipy import stats
import statsmodels.api as sm
import pylab as py
from statsmodels.stats.weightstats import ztest as ztest
```

# Duration in Minutes Vs Start_end_scan 
## Ho : Means are equal
## Ha : Means are not equal

```python
zscore , pvalue1 = ztest(dataset["Duration in Minutes"] , dataset["start_scan_to_end_scan"] , alternative="two-sided")
print(pvalue1)
if pvalue1<0.5:
    print("Reject Null Hypothesis")
else:
    print("Failed to reject Null Hypothesis")
```

```python
def HigherOutlierPercentage(col):
    median = dataset[col].quantile(0.5)
    lower = dataset[col].quantile(0.25)
    higher = dataset[col].quantile(0.75)
    print(col , "median :" , median ,"lower :" , lower, "higher :" , higher)
    iqr = higher-lower
    outlier_value_higher = higher+1.5*(iqr)
    percentage = len(dataset[dataset[col]>outlier_value_higher])/len(dataset[col])*100
    print(col , "Percentage for Upper Outliers :" ,percentage , end = "\n\n" )
```

```python
def LowerOutlierPercentage(col):
    median = dataset[col].quantile(0.5)
    lower = dataset[col].quantile(0.25)
    higher = dataset[col].quantile(0.75)
    print(col , "median :" , median ,"lower :" , lower, "higher :" , higher)
    iqr = higher-lower
    outlier_value_lower = lower-1.5*(iqr)
    percentage = len(dataset[dataset[col]<outlier_value_lower])/len(dataset[col])*100
    print(col , "Percentage for Lower:" ,percentage , end = "\n\n" )
```

```python
continous_columns = ["start_scan_to_end_scan",
       'actual_distance_to_destination', 'actual_time', 'osrm_time',
       'osrm_distance', 'segment_actual_time', 'segment_osrm_time',
        'Duration','Duration in Minutes']
```
